# Coding project - Tumour segmentation

The image below is a slice of a brain MRI with a large tumour in it. The goal of this project is to create an algorithm to automatically segment the tumour.

Given that the resolution of the image is of 0.115 cm/px in both axis, estimate the area of the tumour (in cmÂ²). 

```python
from scipy.ndimage import distance_transform_edt
from skimage.color import rgb2gray
from skimage.color import rgb2hed
from skimage.morphology import disk
from skimage.feature import canny
import skimage.filters.rank as skr
from skimage.segmentation import mark_boundaries, watershed
from skimage.io import imread
from skimage.feature import peak_local_max
from skimage.measure import label
from skimage.filters import gaussian
from skimage.filters import threshold_otsu
from skimage.filters.rank import gradient
import numpy as np
from skimage.io import imread,imshow
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
im = imread('mri_brain.jpg')[:,:,0]
imshow(im)
```

```python
plt.hist(im.flatten(),255);
```

```python
#1st Try
```

```python
from skimage.morphology import disk
from skimage.segmentation import mark_boundaries, watershed

gradient = skr.gradient(skr.mean(im, disk(4)), disk(1))
#Then, it calculates the gradient of the mean-filtered image using a disk-shaped structuring element with a radius of 1 pixel (disk(1)).
markers_coordinates = [[110,60],# tumor
                       [100, 110]#other
]
markers = np.zeros_like(im)

for i,(row,col) in enumerate(markers_coordinates):
    markers[row,col] = i+1
#It creates a marker image (markers) where the pixels specified by markers_coordinates are set to 1 and 2.
ws = watershed(gradient, markers)
#It applies the watershed algorithm to segment the image based on the gradient and marker information.
plt.figure(figsize=[8,8])
plt.subplot(2,2,1)
plt.imshow(im,cmap=plt.cm.gray);
plt.subplot(2,2,2)
plt.imshow(gradient,cmap=plt.cm.gray);
plt.subplot(2,2,3)
plt.imshow(ws);
plt.subplot(2,2,4)
plt.imshow(mark_boundaries(im,ws));
```

```python
plt.imshow(canny(im, sigma=7),cmap=plt.cm.gray);
```

```python
unique_labels, label_counts = np.unique(ws, return_counts=True)
#returns an array of unique labels present in the watershed segmentation result (ws) and their corresponding counts.
#unique_labels contains the unique labels, and label_counts contains the number of occurrences for each label.
print("The tumor has:", label_counts[0]," pixels")
```

```python
area = label_counts[0] * 0.115**2
print("The Tumor's area is", area, "cm^2" )
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
from skimage.segmentation import mark_boundaries
from skimage.segmentation import flood, flood_fill
from skimage.morphology import binary_closing, disk

binary_mask = im > 165
#It creates a binary mask where pixels with intensity greater than 165 in the input image im are set to True, 
#representing the potential tumor region.
binary_mask = binary_closing(binary_mask, disk(4)) 
#It performs a morphological closing operation on the binary mask to smooth and close small gaps or holes. 
#The disk-shaped structuring element with a radius of 4 pixels is used.
seed_point = (110, 60)
seed_mask = np.zeros_like(im, dtype=bool)
seed_mask[seed_point] = 1
#It defines a seed point (seed_point) for the flood-fill operation and creates a binary mask (seed_mask) with True at the seed point.
segmentation_mask = flood(binary_mask, seed_point, connectivity=2)
#It uses the flood function to perform a flood-fill operation starting from the specified seed point. 
#The result is a segmentation mask indicating the connected region.
final_mask = binary_mask & segmentation_mask

tumor_pixel_count = np.sum(final_mask)

plt.figure(figsize=[8, 8])
plt.subplot(1, 2, 1)
plt.imshow(final_mask, cmap=plt.cm.binary)
plt.subplot(1, 2, 2)
plt.imshow(mark_boundaries(im, final_mask))

plt.show()

print(f"Number of pixels in the tumor region: {tumor_pixel_count}")
```

```python
area_total = tumor_pixel_count*0.115**2
print(area_total,"cm^2")
```
